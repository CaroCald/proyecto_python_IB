```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

data_frame_peliculas = pd.read_csv('data/movies.csv')
data_frame_peliculas
```

```python
data_frame_peliculas_genero = data_frame_peliculas['genres'].value_counts()
```

```python
arreglo_labels =[]
arreglo_valores=[]
for val, cnt in  data_frame_peliculas['genres'].value_counts().iteritems():
    arreglo_labels.append(val)
    arreglo_valores.append(cnt)
```

```python
y_pos = np.arange(len(arreglo_labels[1:10]))
plt.bar(y_pos, arreglo_valores[1:10], align='center', alpha=0.9)
plt.xticks(y_pos, arreglo_labels[1:10], rotation=45)
plt.ylabel('Numero de peliculas')
plt.title('Generos de Peliculas mas realizadas')
plt.savefig("reporte_generos.pdf")
plt.show()
```

# Comparativa del presupuesto y la recaudacion de peliculas

```python
df_recaudacion = pd.read_csv('data/movies.csv', index_col='director_name', usecols=['director_name','gross', 'budget'])
df_final =df_recaudacion[:10]
df_final
```
